```python
import csv
import time
import uuid
import requests
import threading
import sys
from datetime import datetime
from pynput import keyboard, mouse

import pywinctl as pwc


class TestMonitor:
    def __init__(self, csv_file="questions.csv"):
        self.test_id = str(uuid.uuid4())[:8].upper()
        self.start_time = datetime.now()
        self.end_time = self.start_time.timestamp() + 1200  # 20 minutes
        self.ip_history = []
        self.copy_paste_events = []
        self.right_click_events = []
        self.window_switches = []
        self.answers = []
        self.running = True
        self.current_ip = self.get_public_ip()
        self.ip_history.append((datetime.now(), self.current_ip))
        self.last_window = pwc.getActiveWindowTitle()
        
        print(f"Test ID: {self.test_id}")
        print(f"Total Test Time: 20 minutes")
        print(f"Initial IP: {self.current_ip}\n")

    def get_public_ip(self):
        try:
            return requests.get('https://api.ipify.org?format=json', timeout=5).json()['ip']
        except:
            return "Unable to fetch IP"

    def monitor_ip(self):
        while self.running:
            time.sleep(30)
            new_ip = self.get_public_ip()
            if new_ip != self.current_ip and new_ip != "Unable to fetch IP":
                self.ip_history.append((datetime.now(), new_ip))
                self.current_ip = new_ip
                print(f"IP CHANGED → {new_ip}")

    def monitor_window(self):
        while self.running:
            time.sleep(1)
            current = pwc.getActiveWindowTitle()
            if current and current != self.last_window and current.strip():
                self.window_switches.append((datetime.now(), self.last_window, current))
                print(f"WINDOW SWITCHED: {self.last_window} → {current}")
                self.last_window = current

    def timer_thread(self):
        while self.running:
            if time.time() > self.end_time:
                print("\n\nTEST TIME IS UP! (20 minutes reached)")
                self.running = False
                break
            time.sleep(5)

    def on_mouse_click(self, x, y, button, pressed):
        if button == mouse.Button.right and pressed:
            self.right_click_events.append((datetime.now(), (x, y)))
            print(f" Right-click detected at ({x}, {y})")

    def on_key_press(self, key):
        if not hasattr(self, 'pressed_keys'):
            self.pressed_keys = set()
        self.pressed_keys.add(key)

    def on_key_release(self, key):
        if hasattr(self, 'pressed_keys'):
            self.pressed_keys.discard(key)
        try:
            if hasattr(key, 'char'):
                if key.char in ('c', 'C') and (keyboard.Key.ctrl_l in self.pressed_keys or keyboard.Key.ctrl_r in self.pressed_keys):
                    self.copy_paste_events.append((datetime.now(), "COPY"))
                    print("Copy detected")
                elif key.char in ('v', 'V') and (keyboard.Key.ctrl_l in self.pressed_keys or keyboard.Key.ctrl_r in self.pressed_keys):
                    self.copy_paste_events.append((datetime.now(), "PASTE"))
                    print("Paste detected")
        except:
            pass

    def load_questions(self, csv_file):
        questions = []
        try:
            with open(csv_file, 'r', encoding='utf-8') as f:
                reader = csv.DictReader(f)
                print(f"CSV Headers found: {reader.fieldnames}\n")  # Debug info
                
                for row in reader:
                    # Flexible column name matching
                    qid = row.get("qid")
                    question = row.get("text")
                    all_answers = row.get("options")
                    correct = row.get("correct")
                    
                    questions.append({
                        "qid": str(qid).strip(),
                        "text": question.strip(),
                        "all_answers": all_answers.strip(),
                        "correct": correct.strip()
                    })
            
            print(f"Loaded {len(questions)} questions successfully.\n")
            # Show first question for verification
            if questions:
                print("Sample of first question:")
                print(f"QID: {questions[0]['qid']}")
                print(f"Text: {questions[0]['text'][:100]}...\n")
            
            return questions
        except Exception as e:
            print(f"Error reading CSV file: {e}")
            sys.exit(1)

    def run_test(self, questions):
        print("\n TEST STARTED (Total 20 minutes) \n")
        
        threading.Thread(target=self.monitor_ip, daemon=True).start()
        threading.Thread(target=self.monitor_window, daemon=True).start()
        threading.Thread(target=self.timer_thread, daemon=True).start()
        
        self.pressed_keys = set()
        kb_listener = keyboard.Listener(on_press=self.on_key_press, on_release=self.on_key_release)
        ms_listener = mouse.Listener(on_click=self.on_mouse_click)
        kb_listener.start()
        ms_listener.start()

        for q in questions:
            if not self.running:
                break
                
            print(f"\n{'='*60}")
            print(f"Question {q['qid']}: {q['text']}")
            if q['all_answers']:
                print(f"\nAll Answers / Options:\n{q['all_answers']}")
            print(f"{'='*60}")
            
            print("Enter your answer below:")
            
            while self.running:
                try:
                    user_answer = input(">>> ").strip()
                    if user_answer or not self.running:
                        break
                except:
                    if not self.running:
                        user_answer = "[TIME_UP]"
                        break
                time.sleep(0.2)
            
            if not user_answer:
                user_answer = "[NO_ANSWER]"
            
            is_correct = user_answer.lower() == q["correct"].lower() if user_answer not in ("[TIME_UP]", "[NO_ANSWER]") else False
            
            self.answers.append({
                "question_id": q["qid"],
                "question": q["text"],
                "user_answer": user_answer,
                "correct_answer": q["correct"],
                "is_correct": is_correct,
                "timestamp": datetime.now().isoformat()
            })
            
            print(f"→ {'Correct ✅' if is_correct else 'Incorrect ❌'}\n")
        
        self.running = False
        kb_listener.stop()
        ms_listener.stop()
        
        self.save_csv()
        self.print_summary()

    def save_csv(self):
        filename = f"test_results_{self.test_id}.csv"
        with open(filename, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow([
                "Test ID", "Question ID", "Question", "User Answer",
                "Correct Answer", "Is Correct", "IP Changed",
                "Copy Paste Used", "Right Clicks Count", "Window Switches Count",
                "Test Duration (min)"
            ])
            
            ip_changed = "Yes" if len(self.ip_history) > 1 else "No"
            cp_used = "Yes" if self.copy_paste_events else "No"
            right_clicks = len(self.right_click_events)
            window_switches = len(self.window_switches)
            duration = round((datetime.now() - self.start_time).total_seconds() / 60, 2)
            
            for ans in self.answers:
                writer.writerow([
                    self.test_id, ans["question_id"], ans["question"],
                    ans["user_answer"], ans["correct_answer"],
                    "Yes" if ans["is_correct"] else "No",
                    ip_changed, cp_used, right_clicks, window_switches, duration
                ])
        
        print(f"\n✅ Results saved to: {filename}")

    def print_summary(self):
        print("\nTEST SUMMARY")
        print(f"Test ID: {self.test_id}")
        print(f"Total Time Used: {round((datetime.now() - self.start_time).total_seconds() / 60, 1)} minutes")
        print(f"IP Changed: {'Yes' if len(self.ip_history) > 1 else 'No'}")
        print(f"Copy/Paste Used: {'Yes' if self.copy_paste_events else 'No'}")
        print(f"Right-clicks: {len(self.right_click_events)}")
        print(f"Window Switches: {len(self.window_switches)}")
        score = sum(1 for a in self.answers if a['is_correct'])
        print(f"Score: {score}/{len(self.answers)}")
```

```python
if __name__ == "__main__":
    csv_file = input("Enter path to questions CSV file (default: questions.csv): ").strip()
    
    monitor = TestMonitor(csv_file)
    questions = monitor.load_questions(csv_file)
    
    if questions:
        monitor.run_test(questions)
    else:
        print("No questions loaded. Exiting.")
```
